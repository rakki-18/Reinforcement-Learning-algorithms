```python
import gym
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
from env.blackjack import BlackjackEnv
from plots.value_function import value_function
```

```python
# getting the environment 
env = BlackjackEnv()
```

```python
def monte_carlo(policy,env,num_episodes,discount_factor):
    """ function returns the value function of states given the number of episodes 
    using monte carlo search"""
    
    # dictionaries to map different states with its total reward, count and value function
    total_sum  = defaultdict(float)
    total_count = defaultdict(float)
    V = defaultdict(float)
    
    # we will repeat the process num_episodes times
    for i in range(num_episodes):
        # list which stores the state and reward in each step of that episode
        episodes  = []
        # starting from a random state of the environment
        state = env.reset()
        # we will roll out one episode of actions 
        for j in range(100):
            # taking a suitable action for that state based on our policy
            action = policy(state)
            # getting the reward and the next state from the environment for the action we took
            next_state, reward,done,_ = env.step(action)
            # store it in our list
            episodes.append((state,reward))
            state = next_state
            # stop once we are done 
            if done:
                break
        # storing all the states we traversed in that episode
        states_episode = set([tuple(x[0]) for x in episodes])
        
        # iterating through all states in that episode to update its value function
        for state in states_episode:
            # finding the occurence of the particular state in our episode 
            first_occurence = next(j for j,x in enumerate(episodes) if x[0]==state)
            # total reward for that state will be the discounted reward obtained in the subsequent states
            G = sum([x[1]*(discount_factor**i) for i,x in enumerate(episodes[first_occurence:])])
            # updating sum and count of that state
            total_sum[state] +=G
            total_count[state]+=1.0
            # updating value function of that state as the mean of reward obtained in different episodes
            V[state] = total_sum[state] / total_count[state]
        
    return V
```

```python
def policy(state):
    # policy followed is sticking if score is >=20, otherwise we will always hit
    score, score_dealer,useable_ace = state
    if(score>=20):
        return 0
    return 1;

 
```

```python
# finding value function for the given policy by running 10000 episodes
V = monte_carlo(policy,env,10000,1)
value_function(V,title = "10000 episodes")
# finding value function by running 500000 episodes
V = monte_carlo(policy,env,500000,1)
value_function(V,title = "500000 episodes")
```
