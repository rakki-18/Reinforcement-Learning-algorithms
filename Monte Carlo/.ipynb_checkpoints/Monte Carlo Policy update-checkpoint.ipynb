```python
import gym
from env.blackjack import BlackjackEnv 
from plots.value_function import value_function # for plotting the value function
import numpy as np
from collections import defaultdict
import matplotlib
```

```python
env = BlackjackEnv() # getting the environment
```

```python
def greedy_policy(state,Q,nA,epsilon):
    """given the q values of every state,action pair and the current state we are in, 
       this function returns a greedy policy to be followed for that state """
    
    # initialising every action with a probability of epsilon/n for exploration
    A = np.ones(nA)*epsilon / nA
    # finding the best action from the q values
    best_action = np.argmax(Q[state])
    # giving an extra probability of 1-epsilon for that action
    A[best_action] += (1-epsilon)
    
    return A
    
```

```python
def monte_carlo_policy_update(env,epsilon,num_episodes,discount_factor):
    """given the environment the function returns 
       the q values of every state,action pair by performing monte carlo epsilon greedy 
       policy update"""
    
    # dictionaries to store the sum and count of state,action pair
    total_sum = defaultdict(float)
    total_count = defaultdict(float)
    # dictionary storing q value of state,action pair
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    
    # number of actions in the environment 
    nA = env.action_space.n
    
    # repeating the process num_episodes times
    for i in range(num_episodes):
        # stores the state,action and reward we get in each episode 
        episodes = []
        # starting from a random state
        state = env.reset()
        # we will roll out one episode of actions
        for j in range(100):
            # finding probability of taking each action in the given state
            probs = greedy_policy(state,Q,nA,epsilon)
            # taking a non-uniform action based on the probabilities we have found 
            action = np.random.choice(np.arange(nA),p = probs)\
            # getting the reward and observations from the environment
            next_state,reward,done,_  = env.step(action)
            # appending the state,action, reward to episodes
            episodes.append((state,action,reward))
            # if we reach a terminal state, we will stop
            if done:
                break
            state = next_state
        # storing all the state and actions we have visited in our episode    
        state_and_action_in_episode = set([(tuple(x[0]), x[1]) for x in episodes])
        
        # iterating through the states and actions to update  q values
        for state,action in state_and_action_in_episode:
            sa_pair = (state,action)
            # finding the first occurrence of that state,action pair
            first_occurrence = next(j for j,x in enumerate(episodes) if x[0]==state and x[1]==action)
            # total reward for that state,action pair will be the discounted reward of the subsequent states
            G = sum([x[2]*(discount_factor**j) for j,x in enumerate(episodes[first_occurrence:])])
            # updating sum and count of that state,action pair
            total_sum[sa_pair]+=G
            total_count[sa_pair]+=1.0
            # updating q value of the state,action pair as mean of reward obtained in different episodes
            Q[state][action] = total_sum[sa_pair]/ total_count[sa_pair]
    return Q
```

```python
Q = monte_carlo_policy_update(env,0.1,500000,1)
```

```python
V = defaultdict(float)
for state,action in Q.items():
    # value function for each state will be the maximum action_value of all its actions
    best_action_value = np.max(action) 
    V[state] = best_action_value\
# plotting the value function
value_function(V,title="500000 episodes")
```
